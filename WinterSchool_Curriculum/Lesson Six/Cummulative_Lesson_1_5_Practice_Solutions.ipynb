```python
## USE THIS SPACE TO PASTE YOUR SOLUTIONS FOR THE ABOVE EXERCISE.
```

```python
def count_ocurrences(array):
  dictionary = {}
  for item in array:
    if not item in dictionary:
      dictionary[item] = 0
    dictionary[item] = dictionary[item] + 1

  return dictionary
```

```python
def count_ocurrences(array): # Additional functionality
  dictionary = {}
  for item in array:
    if not item in dictionary:
      dictionary[item] = 0
    dictionary[item] = dictionary[item] + 1

  unique_values = [key for key, value in dictionary.items() if value == 1]

  return (dictionary, unique_values)
```

```python
def random_password_generator( length_of_password=8, use_special = True):
  import numpy as np
  import random
  import string
  password = ""
  symbols = "!@#$%^&*()_?"

  for i in range(length_of_password):
    val = random.choice(np.arange(3 if use_special else 2))
    if val == 0:
      password = password + str(random.choice(string.ascii_letters))
    elif val == 1:
      password = password + str(random.randint(0,10))
    elif val == 2:
      password = password + str(random.choice(symbols))
  return password
```

```python
# Additional functionality
def random_password_generator( length_of_password=8, use_special = True):
  import numpy as np
  import random
  import string
  password = ""
  symbols = "!@#$%^&*()_?"

  for i in range(length_of_password):

    val = random.choice(np.arange(3 if use_special else 2))
    if val == 0:
      password = password + str(random.choice(string.ascii_letters))
    elif val == 1:
      password = password + str(random.randint(0,9))
    elif val == 2:
      password = password + str(random.choice("!@#$%^&*()_?"))

  if len(password) < 8:
    return "{} the password is weak".format(password)
  else:
    characters = list(password)
    count_upper = len([x for x in characters if x.isupper()])
    count_lower = len([x for x in characters if x.islower()])
    count_num = len([x for x in characters if x.isdigit()])
    count_special = len([x for x in characters if x in symbols])

    if all([count_upper,count_lower, count_special, count_num]) > 0:
      return "{} the password is strong".format(password)
    else:
      return "{} the password is weak".format(password)
  return password
```

```python
for i in range(1,16):
  password = random_password_generator(i, True)
  print(password)
```

- To calculate probability use: (event frequency)/(sample space). Our sample space for two six-sided dice is 36. For this you will need a sample outcome table similar to the one [found in this article](https://www.thoughtco.com/probabilities-of-rolling-two-dice-3126559)

```python
#Additional functionality
def two_cube_dice():
  import numpy as np
  # https://numpy.org/doc/stable/reference/random/generated/numpy.random.randint.html

  # Generate Matrix of All possible events
  events = np.zeros((6,6))
  for i in range(1,7):
    for j in range(1,7):
      events[i-1,j-1] = i+j

  # Count Each event
  unique, counts = np.unique(events, return_counts=True)

  # Make a map when key is the event and the value is number of occurrences
  dictionary = {unique[i]: counts[i] for i in range(len(counts))}

  # Roll Dice
  roll = tuple(np.random.randint(1,7, size=2))

  # Calculate probablity by taking sum of roll and match to map defined above
  probability = ((dictionary[roll[0]+roll[1]])/events.size)*100
  return roll, f"probability of getting {roll[0]+roll[1]} is {round(probability, 1)}%"
```

- Do smokers tip more than non-smokers in average? Calculate this relative to the total bill and percentage tipped. Plot smoker vs non-smoker data.

```python
## USE THIS SPACE TO PASTE YOUR SOLUTIONS FOR THE ABOVE EXERCISE.
```

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
```

```python
df
```

```python
tip_day_df = df.groupby("day")[['tip', "total_bill"]].mean()
```

```python
tip_day_df["tip_percentage_of_bill"] = (tip_day_df['tip']	 / tip_day_df['total_bill']	) * 100
```

```python
tip_day_df
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
for i in range(len(tip_day_df.index.to_list())):
  X = tip_day_df.iloc[i].total_bill
  Y = tip_day_df.iloc[i].tip_percentage_of_bill
  line = ax.scatter(X, Y, color='b')
  # Annotation
  ax.annotate(tip_day_df.index.to_list()[i], xy =(X, Y),
                  xytext =(X+0.1, Y))
plt.xlabel("Total Bill")
plt.ylabel("Tip Percentage");
```
