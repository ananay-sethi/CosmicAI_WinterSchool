```python
list_planets = []
with open("planetary_data_2.txt", 'r') as file:
  next(file)
  for line in file:
    planet = line.strip().replace('\t', '').split(",")
    list_planets.append(planet)
print(list_planets)
print("Planet Info: ", list_planets[int(input("User Index Selection: "))])
```
