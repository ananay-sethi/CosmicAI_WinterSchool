## Exercise 5:

Use the provided code below to complete the exercise.

-Compute the rank of M.

-Compute the SVD decomposition: U, S, Vt.

-Reconstruct M from the SVD components.

- Compute the Frobenius norm of the reconstruction error.

```python
M = np.array([
    [2, 0, 1],
    [1, 3, 4],
    [0, 1, 2]
])

## USE THIS SPACE TO PASTE YOUR SOLUTIONS FOR THE ABOVE EXERCISE.
```

```python
#Solution

M = np.array([
    [2, 0, 1],
    [1, 3, 4],
    [0, 1, 2]
])

# 1. Rank
rank_M = np.linalg.matrix_rank(M)

# 2. Compute SVD
U, S, Vt = np.linalg.svd(M)

# 3. Reconstruction
S_matrix = np.zeros_like(M, dtype=float)
np.fill_diagonal(S_matrix, S)
M_reconstructed = U @ S_matrix @ Vt

# 4. Reconstruction error
error = np.linalg.norm(M - M_reconstructed, ord='fro')

rank_M, U, S, Vt, error
```

```python
Q = np.array([
    [6, -1],
    [2,  3]
])

# 1. Eigenvalues & vectors
vals, vecs = np.linalg.eig(Q)

# 2. Diagonalization
P = vecs
D = np.diag(vals)
Pinv = np.linalg.inv(P)

PDPinv = P @ D @ Pinv   # reconstruct Q

# 3. Direct Q^5
Q5_direct = np.linalg.matrix_power(Q, 5)

# 4. Eigen-decomposition Q^5
D5 = np.diag(vals**5)
Q5_eig = P @ D5 @ Pinv

# 5. Compare
compare = np.allclose(Q5_direct, Q5_eig)

vals, vecs, PDPinv, Q5_direct, Q5_eig, compare
```
